# Generating sound waves

How to generate a sine, sawtooth, square and triangular wave using python
given:
* sf - sampling frequency
* secs - the duration of the signal in seconds
* freq - the frequency of the wave to produce

The sounds can be tested using the [Online Tone Generator](http://onlinetonegenerator.com/)

```python
# pip3 install sounddevice
import sounddevice as sd

import numpy as np
import matplotlib.pyplot as plt
```

```python
# SINE WAVE
# sf: sampling frequency
# secs: duration
# freq: frequency of output signal
def gen_sin(sf, secs, freq):
    nsamples = secs* sf # int(secs/sf)
    arr = np.zeros(nsamples)
    #dt = 1/sf
    for i in range(len(arr)):
        arr[i] = np.sin((i/sf)*freq*(2*np.pi))
    return arr
arr = gen_sin(8000, 5, 440)
sd.play(arr, samplerate = 8000, blocking = True)
plt.plot(arr[0:40])
```

```python
def gen_saw(sf, secs, freq):
    nsamples = secs* sf # int(secs/sf)
    arr = np.zeros(nsamples)
    #dt = 1/sf
    y = 0
    dy = 2 * freq/sf
    for i in range(nsamples):
        #pos = int(i*freq/sf)  % freq
        #y = ( i * 2 * freq/sf
        if y >= 1: y = -1
        arr[i] = y
        y += dy
    return arr
arr = gen_saw(8000, 5, 440)
sd.play(arr, samplerate = 8000, blocking = True)
plt.plot(arr[0:40])
```

```python
def gen_sqr(sf, secs, freq):
    nsamples = secs* sf # int(secs/sf)
    arr = np.zeros(nsamples)
    #dt = 1/sf
    y = 0
    dy = 2 * freq/sf
    for i in range(nsamples):
        if y >= 1: y = -1
        arr[i] = 1 if y>=0 else -1
        y += dy
    return arr
arr = gen_sqr(8000, 5, 440)
sd.play(arr, samplerate = 8000, blocking = True)
plt.plot(arr[0:40])
```

```python
def gen_tri(sf, secs, freq):
    nsamples = secs* sf # int(secs/sf)
    arr = np.zeros(nsamples)
    #dt = 1/sf
    y = 0
    dy = 2 * freq/sf
    for i in range(nsamples):
        if y>= 2: y = 0
        if y < 1: 
            z = y
        else:
            z = 2-y
        arr[i] = z
        y += dy
    return arr
arr = gen_tri(8000, 5, 440)
sd.play(arr, samplerate = 8000, blocking = True)
plt.plot(arr[0:40])
```
